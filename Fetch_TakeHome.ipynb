```python
# Function to check for missing values and duplicates
def data_quality_check(df, name):
    print(f"--- {name} Data Quality Check ---")
    print("Missing Values:\n", df.isnull().sum())
    print("\nDuplicate Rows:", df.duplicated().sum())
    print("\nData Types:\n", df.dtypes)
    print("\nUnique Value Counts:\n", df.nunique())
    print("\n" + "="*50 + "\n")

# Perform data quality check
data_quality_check(products, "Products")
data_quality_check(transactions, "Transactions")
data_quality_check(users, "Users")
```

```python
# Function to calculate missing value percentages
def calculate_missing_percentage(df, df_name):
    missing_percent = df.isnull().mean() * 100
    missing_percent = missing_percent[missing_percent > 0].sort_values(ascending=False)
    print(f"\nMissing Values Percentage in {df_name} Dataset:\n{missing_percent.round(2)}\n")

# Calculate and print missing value percentages for each DataFrame
calculate_missing_percentage(products, "Products")
calculate_missing_percentage(transactions, "Transactions")
calculate_missing_percentage(users, "Users")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to visualize missing values
def plot_missing_values(df, name):
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values(ascending=False)

    if not missing.empty:
        plt.figure(figsize=(10, 5))
        sns.barplot(x=missing.index, y=missing.values, palette="viridis")
        plt.xticks(rotation=45)
        plt.xlabel("Column Name")
        plt.ylabel("Missing Values Count")
        plt.title(f"Missing Values in {name} Dataset")
        plt.show()

# Plot missing values and mixed data types for each dataset
plot_missing_values(products, "Products")
plot_missing_values(transactions, "Transactions")
plot_missing_values(users, "Users")
```

## What are the top 5 brands by receipts scanned among users 21 and over?

This SQL query retrieves the top 5 most purchased brands by users who are 21 years or older, based on the number of receipts scanned.

1. Joins three tables:
transactions (t): Contains transaction details.
users (u): Provides user demographics (e.g., birth date).
products (p): Holds product and brand information.
2. Filters users who are at least 21 years old using timestampdiff(year, u.birth_date, curdate()) >= 21.
3. Ensures only products with a valid brand name (p.brand is not null) are considered.
4. Groups data by brand and counts the number of receipts (count(t.receipt_id)) to measure brand popularity.
5. Sorts brands in descending order of receipt count to find the most frequently purchased ones.
6. Limits the results to the top 5 brands.

```python
# SQL query

result = spark.sql(
        """
select p.brand,
       count(t.receipt_id) as receipt_count
from transactions t
join users u on t.user_id = u.id
join products p on t.barcode = p.barcode
where timestampdiff(year, u.birth_date, curdate()) >= 21
      and p.brand is not null
group by p.brand
order by receipt_count desc
limit 5;

""")

result.toPandas()
```

At first, I used LIMIT 5 to retrieve the top 5 brands from the table. However, I noticed that some brands have identical receipt counts. To address this, I applied the RANK() function to assign the same rank to brands with equal receipt counts. Determining whether these brands all fall within the top 5 will depend on the stakeholder's definition.

```python
# SQL query

result = spark.sql(
    """
with ranked_brands as (
    select p.brand,
           count(t.receipt_id) as receipt_count,
           rank() over (order by count(t.receipt_id) desc) as rnk
    from transactions t
    join users u on t.user_id = u.id
    join products p on t.barcode = p.barcode
    where timestampdiff(year, u.birth_date, curdate()) >= 21
          and p.brand is not null
    group by p.brand
)

select brand,
       receipt_count,
       rnk
from ranked_brands
where rnk <= 5
order by rnk;

""")

result.toPandas()
```

## What are the top 5 brands by sales among users that have had their account for at least six months?

This query identifies the top 5 brands by total sales among users who have had an account for at least six months.

1. Joins three tables: transactions (t): Contains purchase details.
users (u): Provides user information (including account creation date).
products (p): Holds product and brand details.
2. Filters users who created their account at least six months ago using: u.created_date <= add_months(current_date(), -6)
This ensures that only long-term users are included in the analysis.

3. Excludes transactions where the brand is missing (p.brand is not null).

4. Calculates total sales per brand using sum(t.final_sale).

5. Ranks brands in descending order of total sales and selects the top 5 brands (limit 5).

```python
# SQL query

result = spark.sql(
    """
select p.brand,
       sum(t.final_sale) as total_sales
from transactions t
join users u on t.user_id = u.id
join products p on t.barcode = p.barcode
where u.created_date <= add_months(current_date(), -6)
      and p.brand is not null
group by p.brand
order by total_sales desc
limit 5;

""")

result.toPandas()
```

## Who are Fetchs power users?

I define final_sale as the total amount spent by a user per transaction, representing the final purchase value after any discounts or adjustments. This metric is used to aggregate user spending and identify Fetchs highest-value customers. The query applies LIMIT 10 to optimize performance and focus on the top 10 spenders, providing key business insights. If a broader analysis is needed, the LIMIT can be adjusted accordingly.

This query identifies the top 10 highest-spending users based on their total purchases:

1. Aggregates total spending per user: Groups transactions by user_id and sums final_sale to calculate total_spent, filtering out null or empty values for accuracy.
2. Ranks users by total spending (order by total_spent desc) to prioritize the highest spenders.
3. Limits the results to the top 10 users (limit 10) to highlight Fetchs most valuable customers.

```python
# SQL query

result = spark.sql(
    """
select user_id,
      sum(final_sale) as total_spent
from transactions
where final_sale is not null and final_sale != ''
group by user_id
order by total_spent desc
limit 10;

""")

result.toPandas()
```
